```python
df_category = pd.read_csv('donnees/product_category_name_translation.csv')
df_category.head()
translator = dict()
[translator.update({df_category['product_category_name'][i]:df_category['product_category_name_english'][i]}) for i in range(df_category.shape[0])]
translator
```

```python
df_products = pd.read_csv('donnees/olist_products_dataset.csv')
df_products["product_category_name"] = df_products["product_category_name"].replace(translator)
df_products["product_super_category_name"] = df_products["product_category_name"].replace(super_categories)
df_products["dimension"] = df_products[['product_length_cm','product_height_cm','product_width_cm']].apply(lambda dim : dim[0]*dim[1]*dim[2], axis=1)
df_products["product_name_description_lenght"] = df_products[['product_name_lenght','product_description_lenght']].apply(lambda l : l[0]+l[1], axis=1)
df_products.drop(columns=['product_name_lenght','product_description_lenght','product_category_name','product_width_cm','product_height_cm','product_length_cm'],inplace=True)
df_products['product_super_category_name'] = df_products['product_super_category_name'].replace('portateis_cozinha_e_preparadores_de_alimentos','Technology')
df_products
```

```python
df_items = pd.read_csv('donnees/olist_order_items_dataset.csv')
df_items = df_items.drop(['shipping_limit_date'],axis=1)
df_items
```

```python
df_product_items = df_products.merge(df_items,on='product_id')
df_product_items
```

```python
df_geo = pd.read_csv('donnees/olist_geolocation_dataset.csv')
df_geo = df_geo.groupby('geolocation_zip_code_prefix').agg({'geolocation_lat':'mean','geolocation_lng':'mean'}).reset_index()
df_geo
```

```python
df_sellers = pd.read_csv('donnees/olist_sellers_dataset.csv')
df_sellers.drop(columns=['seller_city','seller_state'],inplace=True)
df_sellers
```

```python
df_geo_sellers = df_sellers.merge(df_geo, left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix')
df_geo_sellers
```

```python
df_items_geo_sellers = df_items.merge(df_geo_sellers, on='seller_id').rename(columns={'geolocation_lat':'seller_lat','geolocation_lng':'seller_lng'})
df_items_geo_sellers
```

```python
df_products_items_geo_sellers = df_products.merge(df_items_geo_sellers, on='product_id').drop(columns=['seller_zip_code_prefix','geolocation_zip_code_prefix'])
df_products_items_geo_sellers
```

```python
df_payments = pd.read_csv('donnees/olist_order_payments_dataset.csv')
df_payments = df_payments[df_payments.payment_type != 'not_defined']
df_payments = df_payments[df_payments.payment_sequential <=4]
df_payments = df_payments[df_payments.payment_installments !=0]
df_payments
```

```python
df_reviews = pd.read_csv('donnees/olist_order_reviews_dataset.csv')
df_reviews.review_comment_message = df_reviews.review_comment_message.apply(lambda s : 0 if str(s)=='nan' else len(s))
df_reviews.review_comment_title = df_reviews.review_comment_title.apply(lambda s : 0 if str(s)=='nan' else len(s))
df_reviews["nb_char_comment"] = df_reviews[['review_comment_message','review_comment_title']].apply(lambda x: x[0]+x[1], axis=1)
df_reviews = df_reviews.drop(['review_comment_title','review_comment_message','review_creation_date','review_answer_timestamp'],axis=1)
df_reviews.head()
```

```python
df_reviews_payments = df_reviews.merge(df_payments, on='order_id')
df_reviews_payments
```

```python
df_orders = pd.read_csv('donnees/olist_orders_dataset.csv')
df_orders = df_orders.drop(columns=['order_delivered_carrier_date','order_delivered_customer_date','order_status','order_approved_at'])
df_orders['order_estimated_delivery_days'] = df_orders[['order_purchase_timestamp','order_estimated_delivery_date']].apply(lambda d: dt.datetime.strptime(d[1], '%Y-%m-%d %H:%M:%S') - dt.datetime.strptime(d[0], '%Y-%m-%d %H:%M:%S'), axis=1)
df_orders['recense'] = (pd.to_datetime('2018-12-31')-pd.to_datetime(df_orders['order_purchase_timestamp'])).apply(lambda d: d.days)
df_orders['order_estimated_delivery_days'] = df_orders['order_estimated_delivery_days'].apply(lambda d : d.days)
df_orders = df_orders.drop(['order_estimated_delivery_date','order_purchase_timestamp'],axis=1)
df_orders
```

```python
df_customers = pd.read_csv('donnees/olist_customers_dataset.csv')
df_customers = df_customers.drop(columns=['customer_state','customer_city'])
df_customers
```

```python
# df_orders_customers = df_orders.merge(df_customers, on='customer_id')
df_orders_customers = df_orders.merge(df_customers, right_on='customer_id', left_on='customer_id')
df_orders_customers
```

```python
df_orders_customers_reviews_payments = df_reviews_payments.merge(df_orders_customers, on='order_id')
df_orders_customers_reviews_payments
```

```python
df_orders_geo_customers_reviews_payments = df_orders_customers_reviews_payments.merge(df_geo, left_on ='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix').rename(columns={'geolocation_lat':'customer_lat','geolocation_lng':'customer_lng'}).drop(columns=['customer_zip_code_prefix','geolocation_zip_code_prefix'])
df_orders_geo_customers_reviews_payments
```

```python
df_merged = df_orders_geo_customers_reviews_payments.merge(df_products_items_geo_sellers, on='order_id')
df_merged

customer_coordinates = df_merged[['customer_lat', 'customer_lng']].values
seller_coordinates = df_merged[['seller_lat', 'seller_lng']].values
df_merged['distance'] = [distance.distance(coord1, coord2).km for coord1, coord2 in zip(customer_coordinates, seller_coordinates)]
df_merged = df_merged.drop(columns=['customer_lat','customer_lng','seller_lat','seller_lng'])
df_merged
```

```python
def mode(v):
    if type(v)==str:
        return [v][0]
    else:
        try:
            return list(v)[0]
        except:
            return 'erreur'


agg = {'order_id':'nunique','order_item_id':'count', 'product_id':'nunique','seller_id':'nunique','price':'mean','freight_value':'mean', 'product_photos_qty':'mean', 'product_weight_g':'mean','product_super_category_name':pd.Series.mode, 'dimension':'mean', 'product_name_description_lenght':'mean', 'review_score':'mean','nb_char_comment':'mean', 'payment_sequential':'mean', 'payment_installments':'mean', 'payment_type':pd.Series.mode, 'order_estimated_delivery_days':'mean', 'distance':'mean', 'recense':'max'}

df_final = df_merged.groupby('customer_unique_id').agg(agg).reset_index().rename(columns={'order_id':'nb_order','order_item_id':'nb_items','seller_id':'nb_sellers','product_id':'nb_product', 'order_purchase_timestamp':'recense'}).drop(columns='customer_unique_id')

df_final['product_super_category_name'] = df_final['product_super_category_name'].apply(lambda v : mode(v))
df_final['payment_type'] = df_final['payment_type'].apply(lambda v : mode(v))
# df_final['product_super_category_name'] = df_final['product_super_category_name'].replace('portateis_cozinha_e_preparadores_de_alimentos','Technology')
# df_final['recense'] = (pd.to_datetime('2018-12-31')-pd.to_datetime(df_final['recense'])).apply(lambda d: d.days)
# df_final['order_estimated_delivery_time'] = df_final['order_estimated_delivery_time'].apply(lambda d : d.days)
df_final = df_final.dropna()
df_final
```
